```python
import json
import os
import numpy as np
from nltk.tokenize import word_tokenize
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction

os.chdir('/collection/ka2khan/thesis/Dialog_Generation')
print(os.getcwd())


with open('experiments/DailyDialog_TACGAN_wo_EncDec_Backprop/valid_output_77.json') as f_obj:
    valid_data = json.load(f_obj)

with open('experiments/t5/generated_responses.txt') as f_obj:
    gen_responses = f_obj.readlines()
    
bleu_scores = []
for index in range(len(valid_data)):
    o_dict = valid_data[index]
    response = gen_responses[index]
    pred_response = response[response.find('->')+2:]
    gt_response = o_dict['response']
    
    bleu_scores.append(sentence_bleu([gt_response], word_tokenize(pred_response), smoothing_function=SmoothingFunction().method7,
                                        weights=[1./3, 1./3, 1./3]))

print(np.mean(bleu_scores))
    
    
```

```python
import json
import os
import numpy as np
from nltk.tokenize import word_tokenize

os.chdir('/collection/ka2khan/thesis/Dialog_Generation')
print(os.getcwd())

with open('experiments/t5_w_keywords_wo_encdec_backprop/test_output_5.json') as f_obj:
    output = json.load(f_obj)

gt_res_tokens = []
pred_res_tokens = []
gt_lengths = []
pred_lengths = []
for o_dict in output:
    gt_lengths.append(len(o_dict['response']))
    gt_res_tokens.extend(o_dict['response'])
    pred_tokens = word_tokenize(o_dict['generated_responses'][0])
    pred_lengths.append(len(pred_tokens))
    pred_res_tokens.extend(pred_tokens)

print(f'Ground Truth TTR: {(len(set(gt_res_tokens)) / len(gt_res_tokens)) * 100}')
print(f'Predicted TTR: {(len(set(pred_res_tokens)) / len(pred_res_tokens)) * 100}')
print(f'Ground Truth ASL: {np.mean(gt_lengths):.2f}')
print(f'Ground Truth ASL: {np.mean(pred_lengths):.2f}')
```

```python
import math
import numpy as np
import torch
from transformers import GPT2Tokenizer, GPT2Config, GPT2LMHeadModel
from tqdm import tqdm

def calculatePerplexity(sentence, model, tokenizer):
    input_ids = torch.tensor(tokenizer.encode(sentence)).unsqueeze(0) 
    input_ids = input_ids.to('cpu')
    with torch.no_grad():
        outputs = model(input_ids, labels=input_ids)
    loss, logits = outputs[:2]
    return math.exp(loss)

import json
import os
import numpy as np
from nltk.tokenize import word_tokenize

os.chdir('/collection/ka2khan/thesis/Dialog_Generation')
print(os.getcwd())

with open('experiments/t5_w_keywords_wo_encdec_backprop/test_output_5.json') as f_obj:
    output = json.load(f_obj)
    
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
config = GPT2Config.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2', config=config)
model.eval()
    
gt_ppl = []
pred_ppl = []
count_nan = 0
for o_dict in tqdm(output):
    ppl = calculatePerplexity(' '.join(o_dict['response']), model, tokenizer)
    if not math.isnan(ppl):
        gt_ppl.append(ppl)
    else:
        count_nan += 1
    ppl = calculatePerplexity(o_dict['generated_responses'][0], model, tokenizer)
    pred_ppl.append(ppl)

print(f'Ground Truth PPL: {np.mean(gt_ppl):.2f}')
print(f'Predicted PPL: {np.mean(pred_ppl):.2f}')
print(f'Total nan values: {count_nan}')
      
```

```python
import os
import sys
import torch
import logging
from torch import nn
from torch.utils.data import TensorDataset, DataLoader

sys.path.append('/collection/ka2khan/thesis/Dialog_Generation')

from models.Neural_LM import Neural_LM
from models.BOW_VAE import BOW_VAE
from models.GAN_Generator import GAN_Generator

from utils import *
from config import config

os.chdir('/collection/ka2khan/thesis/Dialog_Generation')
print(os.getcwd())

log_file = os.path.join('/tmp/process.log')
open(log_file, 'a').close()
    
logging.basicConfig(filename=log_file,
                    filemode='a',
                    format='%(asctime)s %(name)s %(message)s',
                    datefmt='%H:%M:%S',
                    level=logging.INFO)

logger = logging.getLogger(f"Inference ->")

config.dataset = 'DailyDialog'

context_seqs, context_seq_lengths, query_seqs, query_seq_lengths, response_seqs, response_seq_lengths, query_vectors,  response_vectors, vectorizer, index2word = load_gan_dataset(config, logger)
    
config.vocab_size = len(vectorizer.vocabulary_)
    
datasets = {}
dataloaders = {}
for part in config.datasets[config.dataset]['parts']:
        
    datasets[part] = TensorDataset(torch.arange(query_vectors[part].size(0)))
    if part == 'train':
        shuffle = True
    else:
        shuffle = False
            
    dataloaders[part] = DataLoader(datasets[part], batch_size=config.gan_batch_size, shuffle=shuffle, pin_memory=True)

lm_model = load_model(config.lm_path)
vae_model = load_model(config.vae_path)

generator = torch.load('experiments/DailyDialog_TACGAN_wo_EncDec_Backprop/best_model_77.pt')
```

```python
lm_model.eval()
vae_model.eval()
generator.eval()

generator.eval()
vae_model.eval()
lm_model.eval()
with torch.set_grad_enabled(False):
    ref_context_seq = []
    ref_query_seq = []
    ref_response_seq = []
    query_references = []
    response_references = []
    hypotheses = []
    val_recall_bleu = []
    val_precision_bleu = []
    val_tokens_ret = []
    val_total_tokens = []
    for batch in dataloaders['valid']:
        indices = batch[0].cuda()

        context_x = context_seqs['valid'][indices].cuda()
        context_lengths = context_seq_lengths['valid'][indices].cuda()
        query_x = query_seqs['valid'][indices].cuda()
        query_lengths = query_seq_lengths['valid'][indices].cuda()
        response_x = response_seqs['valid'][indices].cuda()
        response_lengths = response_seq_lengths['valid'][indices].cuda()

        query_vecs = query_vectors['valid'][indices].cuda() 
        response_vecs = response_vectors['valid'][indices].cuda() 

        # Obtain context representation
        context_sorted_indices = torch.argsort(context_lengths, descending=True)

        context_x = context_x[context_sorted_indices]
        context_lengths = context_lengths[context_sorted_indices]

        context_h_n, _ = lm_model(context_x, context_lengths)

        context_reverse_indices = torch.argsort(context_sorted_indices)

        context_h_n = context_h_n[context_reverse_indices]

        context_x = context_x[context_reverse_indices]

        # Obtain query representation
        query_sorted_indices = torch.argsort(query_lengths, descending=True)

        query_x = query_x[query_sorted_indices]
        query_lengths = query_lengths[query_sorted_indices]

        query_h_n, _ = lm_model(query_x, query_lengths)

        query_reverse_indices = torch.argsort(query_sorted_indices)

        query_h_n = query_h_n[query_reverse_indices]

        query_x = query_x[query_reverse_indices]

        # Obtain query and response latent codes
        _, _, queries_z = encode_vectors(query_vecs, vae_model, 1.0, num_samples=config.latent_samples)
        
        context_h_n = torch.repeat_interleave(context_h_n, config.latent_samples, dim=0)
        query_h_n = torch.repeat_interleave(query_h_n, config.latent_samples, dim=0)
        
        print(queries_z.size())
        print(context_h_n.size())
        print(query_h_n.size())

        gen_input = torch.cat((context_h_n.detach(), query_h_n.detach(), queries_z), dim=-1)

        predicted_z = generator(gen_input)

        vae_logits = decode_dialog(predicted_z, vae_model)

        recall_bleus, precision_bleus, tokens_ret, total_tokens, refs, hypos = multi_compute_metrics(vae_logits, response_vecs, config.sample_count, vectorizer, config.latent_samples)
        ref_context_seq.extend(indices2words(context_x.detach().cpu().tolist(), index2word))
        ref_query_seq.extend(indices2words(query_x.detach().cpu().tolist(), index2word))
        ref_response_seq.extend(indices2words(response_x.detach().cpu().tolist(), index2word))
        query_references.extend(one_hot_to_bow(query_vecs, vectorizer))
        response_references.extend(refs)
        hypotheses.extend(hypos)
        val_recall_bleu.extend(recall_bleus)
        val_precision_bleu.extend(precision_bleus)
        val_tokens_ret.extend(tokens_ret)
        val_total_tokens.extend(total_tokens)
        break
    
    
    for index in range(len(query_references)):
        print(f'Query: {query_references[index]}')
        for h_index in range(config.latent_samples):
            print(f'Predicted response #{h_index+1}: {hypotheses[(index*config.latent_samples)+h_index]}')

```
